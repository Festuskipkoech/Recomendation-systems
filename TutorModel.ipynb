```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/udemy_course_data.csv')
```

```python
dataset=data[["course_id", "course_title","subject"]]
```

```python
dataset["tags"]=dataset["course_title"]+ dataset["subject"]
```

```python
dataset
```

```python
new_data=dataset.drop(columns=["subject"])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data["tags"].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data['course_title'] == "Ultimate Investment Banking Course"].index[0]
```

```python
distance=sorted(list(enumerate(similarity[2])), reverse=True, key=lambda vector:vector[1])
for i in distance[0:5]:
  print(new_data.iloc[i[0]].course_title)
```

```python
def recommend(title):
  # subject represent dynamic titles users will provide when recommend function is called
  index=new_data[new_data["course_title"] ==title].index[0]
  distance=sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
  for i in distance[0:5]:
    print(new_data.iloc[i[0]].course_title)
```

```python
recommend("Learn and Build using Polymer")
```
